```python
import pandas as pd
import plotly.graph_objects as go
```

```python
df = pd.read_pickle('/content/sample_data/cleaned_pokemon_data.pkl')
# 1. Data Aggregation: Count Pokémon by primary type
primary_type_counts = df['Primary Type'].value_counts()
```

```python
# 2. Color Scheme Mapping
type_colors = {
    'Grass': '#78C850', 'Fire': '#F08030', 'Water': '#6890F0',
    'Bug': '#A8B820', 'Normal': '#A8A878', 'Poison': '#A040A0',
    'Electric': '#F8D030', 'Ground': '#E0C068', 'Fairy': '#EE99AC',
    'Fighting': '#C03028', 'Flying': '#A890F0', 'Psychic': '#F85888',
    'Rock': '#B8A038', 'Ghost': '#705898', 'Ice': '#98D8D8',
    'Dragon': '#7038F8', 'Dark': '#705848', 'Steel': '#B8B8D0',
}
```

```python
# 3. Bar Chart Visualization
fig = go.Figure()

fig.add_trace(
    go.Bar(
        x=primary_type_counts.index,
        y=primary_type_counts,
        marker=dict(color=[type_colors[ptype] for ptype in primary_type_counts.index]),
        name='All Bars in a Figure',
        showlegend=True
    )
)

fig.update_layout(
    title_text='Pokémon Primary Type Distribution by Generation',
    height=300,
    width=800
)
fig.show()

fig = go.Figure()
# Add bars one by one to customize legend and colors
for ptype in primary_type_counts.index:
    fig.add_trace(
        go.Bar(
            x=[ptype],
            y=[primary_type_counts[ptype]],
            marker=dict(color=type_colors.get(ptype, '#000000')),
            name=ptype
        )
    )

fig.update_layout(
    title_text='Pokémon Primary Type Distribution by Generation',
    height=300,
    width=800,
    showlegend=True
)

fig.show()
```

```python
# 4. Bar Chart Configuration
fig.update_layout(
    title_text='Pokémon Primary Type Distribution - Overall',
    xaxis_title='Primary Type',
    yaxis_title='Number of Pokémon',
    height=400,
    width=800,
    showlegend=True
)

# Show the figure
fig.show()
```
